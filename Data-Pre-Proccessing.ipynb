```python
import pandas as pd
import numpy as np
```

# Read Two Data Sets

```python
df_fake = pd.read_csv("Dataset/Fake.csv", index_col=False)
df_true = pd.read_csv("Dataset/True.csv", index_col=False)
```

```python
df_fake.head(5)
```

```python
df_true.head(5)
```

# 1. Encoding Categorical Class:

```python
df_fake["class"] = 0
df_true["class"] = 1
```

```python
df_fake.head()
```

```python
df_true.head()
```

# Merging both datasets into one

```python
print(df_fake.size)
print(df_true.size)
```

```python
df_merge = pd.concat([df_fake, df_true], axis =0 )
print(df_merge.size)
df_merge.head()
```

# Shuffling the rows

```python
df_merge = df_merge.sample(frac=1)
df_merge
```

# 2. Dimension Reduction / Feature Selection:

```python
final_df = pd.DataFrame()
```

```python
# Dropping the index and Date columns (attributes):
final_df['Title'] = df_merge['title']
final_df['Text'] = df_merge['text']
final_df['Class'] = df_merge['class']
```

```python
final_df.reset_index(drop=True, inplace=True)
```

```python
final_df
```

# 3. Data Sampling: --> Stratified Sampling so classes are represented equally

```python
print("Length before sampling: ", len(final_df))
```

```python
# Sampling with a condition:

condition_1 = final_df['Class']==1
condition_0 = final_df['Class']==0

true_index_1 = condition_1[condition_1 == True].index
true_index_0 = condition_0[condition_0 == True].index

subset_1 = final_df[condition_1].sample(n = 500)
subset_0 = final_df[condition_0].sample(n = 500)

# Merge:
merged_final_df = pd.concat([subset_1, subset_0], axis =0).sample(frac=1)
merged_final_df.reset_index(drop=True, inplace=True)
```

```python
merged_final_df
```

```python
print("Length After sampling: ", len(merged_final_df))
```

# 4. Dealing with missing values: Removing Null Values:

```python
print("Size before: ", len(merged_final_df))
```

```python
merged_final_df = merged_final_df.loc[merged_final_df['Text'] != ' ']
```

```python
print("Size after: ", len(merged_final_df))
```

# 5. Removing Unfamiliar Characters:

```python
chars = [';','/', '\\\\', '@', ':', '-', '_', '\'', '\""', '.', 'Ä', 'ô']
for c in chars:
    merged_final_df['Text'] = merged_final_df['Text'].str.replace(c,'')
```
